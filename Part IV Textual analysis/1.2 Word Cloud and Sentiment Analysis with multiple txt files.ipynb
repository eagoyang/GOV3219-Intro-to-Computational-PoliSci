```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import seaborn as sns
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from textblob import TextBlob
import string
import re

import glob
import os
```

### Step 2: Load the CSV File

```python
folder_path = "C:/Users/kaiyang/Desktop/SCMP_articles/"

# Use glob to find all .txt files in that folder
file_paths = glob.glob(os.path.join(folder_path, "*.txt"))

# Create empty lists to store the data
doc_ids = []
article_texts = []

# Loop through each file and read its content
for path in file_paths:
    # Extract the filename without extension (e.g., "article1")
    doc_name = os.path.basename(path).replace(".txt", "")
    
    # Read the article text
    with open(path, "r", encoding="utf-8") as f:
        text = f.read()
    
    # Append to our lists
    doc_ids.append(doc_name)
    article_texts.append(text)

# Create DataFrame (same structure as before, but with multiple rows)
df = pd.DataFrame({'doc_id': doc_ids, 'text': article_texts})

# Check the result
print(f"Successfully loaded {len(df)} articles")
print(df.head())
```

### Step 3: Preprocess the Text

```python
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')

# Initialize Lemmatizer
lemmatizer = WordNetLemmatizer()

def clean_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'\d+', '', text)  # Remove numbers
    text = re.sub(r'[^\w\s]', '', text)  # Remove ALL punctuation (including smart quotes)
    
    tokens = word_tokenize(text)  # Tokenize words
    
    tokens = [word for word in tokens if word not in stopwords.words('english')]  # Remove stopwords
    lemmatized_tokens = [lemmatizer.lemmatize(word) for word in tokens]  # Apply lemmatization
    return " ".join(lemmatized_tokens)

# Apply cleaning function to all titles
df['cleaned_text'] = df['text'].astype(str).apply(clean_text)

# Check cleaned text
print(df[['text', 'cleaned_text']].head())

df.head()
```

### Step 4: Generate a Word Cloud

```python
# Combine all titles into a single string
text = " ".join(df['cleaned_text'])

# Generate word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white', colormap='viridis').generate(text)

# Plot the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud of News Text")
plt.show()
```

### Step 5: Sentiment Analysis

```python
def get_sentiment(text):
    analysis = TextBlob(text)
    return analysis.sentiment.polarity  # Returns value between -1 (negative) and 1 (positive)

# Apply sentiment function
df['sentiment'] = df['cleaned_text'].apply(get_sentiment)

# Check results
print(df[['text', 'sentiment']].head())
```

### Step 6: Visualize Sentiment Distribution

```python
plt.figure(figsize=(10, 6))
plt.hist(df['sentiment'], bins=10, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(x=0, color='red', linestyle='--', linewidth=2, label='Neutral (0)')
plt.xlabel('Sentiment Score')
plt.ylabel('Number of Articles')
plt.title('Distribution of Sentiment Scores Across Articles')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(12, 6))

# Sort by sentiment score (optional - makes trends easier to see)
df_sorted = df.sort_values('sentiment', ascending=False)

# Create bars with colors based on sentiment
colors = ['green' if x > 0 else 'red' if x < 0 else 'gray' for x in df_sorted['sentiment']]
bars = plt.bar(df_sorted['doc_id'], df_sorted['sentiment'], color=colors, edgecolor='black')

# Add a horizontal line at zero (neutral)
plt.axhline(y=0, color='black', linestyle='-', linewidth=0.8)

# Add value labels on top of each bar
for bar, score in zip(bars, df_sorted['sentiment']):
    plt.text(bar.get_x() + bar.get_width()/2, 
             bar.get_height() + (0.02 if score >= 0 else -0.05),
             f'{score:.3f}', 
             ha='center', 
             va='bottom' if score >= 0 else 'top',
             fontsize=9)

# Customize the plot
plt.xlabel('Article Name', fontsize=12)
plt.ylabel('Sentiment Score', fontsize=12)
plt.title('Sentiment Scores for Each News Article', fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)

# Add a legend
from matplotlib.patches import Patch
legend_elements = [Patch(facecolor='green', label='Positive'),
                   Patch(facecolor='gray', label='Neutral'),
                   Patch(facecolor='red', label='Negative')]
plt.legend(handles=legend_elements, loc='upper right')

plt.tight_layout()
plt.show()
```
