Word cloud & word frequency

### Word Cloud

```python
# pip install wordcloud matplotlib
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Example text 
text = "Python is a great language for data analysis. Python is widely used in machine learning and data science. Python is used in Introduction to Computational Political Science offered by the Department of Government and International Affairs"

# generating word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# visualizing word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # close axises
plt.show()
```

### Word Frequency

```python
# pip install nltk
```

```python
from collections import Counter
import nltk
from nltk.corpus import stopwords
import string

# download stopwords
nltk.download('stopwords')

# example text
# text = "Python is a great language for data analysis. Python is widely used in machine learning and data science. Python is used in Introduction to Computational Political Science offered by the Department of Government and International Affairs"

# split and remove puctuation and stopwords
words = text.lower().split()  # split and lowercase
words = [word.strip(string.punctuation) for word in words]  # remove punctuation 
stop_words = set(stopwords.words('english'))  # english stopwords
words = [word for word in words if word not in stop_words and word]  # remove stopwords and empty strings

# calculate word frequency
word_freq = Counter(words)

# return word frequency 
print("word frequency：")
for word, freq in word_freq.most_common():
    print(f"{word}: {freq}")
```
